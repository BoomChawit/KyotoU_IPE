```python
import numpy as np
from matplotlib import pyplot as plt
g = np.loadtxt("../ILAS_PyEng2019/sample_data/signal_data.csv", dtype='float',delimiter=',')
plt.plot(g[0],g[1])

def sin_function(x, a, b): 
    y = a * np.sin(b+x) 
    return y     

from scipy.optimize import curve_fit

opt, cov = curve_fit(sin_function, g[0], g[1])
yfit = sin_function(g[0], *opt)

e = (yfit - g[1])
rmse = np.sqrt(np.sum(e**2)/ len(g[1]))
print(f'RMSE = {round(rmse,3)}')
plt.plot(g[0],yfit, 'r')
```
