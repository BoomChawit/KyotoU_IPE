# Unique

```python
a = [1,2,3,'e','i','o','u',23,34,2,3,1,'o']
len(a)
```

```python
b =[]

if len(b)==0:
        b.append(a[0])
for i in a:
    print(i)
    for j in b:
        print(j)
        if i == j:
            break
        else:
            b.append(i)
            print(b)
            print("end")
print(b)
print(len(b))
```

```python
b=[]
i = 0
j = 0
while i <= len(a):
    while j <= len(b):
        if not a[i] == b[j]:
            j+=1
    i+=1

print(b)
```

```python
a = [1,2,3,'e','i','o','u',23,34,2,3,1,'o']
b = []

for i in a:
    x = True
    for j in b:
        if i==j:
            x = False
            break
            
    if x == True:
        b.append(i)
        
print(b)
```
